```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[50,52,48,64,65,63,65]
```

```python

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('seekly weather')
plt.plot(x,y,color='red', linestyle='-.')
```

```python
plt.plot(x,y,color='red', marker = '+',linestyle='-.')
```

```python
plt.plot(x,y,color='red', marker = '+',linestyle='-.',markersize=20)
```

```python
plt.plot(x,y,color='red', marker = '+',linestyle='-.',markersize=20,linewidth=13)
```

```python
plt.plot(x,y,'g<', alpha=0.5)
```
