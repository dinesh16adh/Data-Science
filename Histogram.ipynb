```python
import matplotlib.pyplot as plt
```

```python
blood_sugar=[113,85,90,99,100,150,149,88,93,155,135,80,77]
plt.hist(blood_sugar,rwidth=0.8);
```

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4);
```

```python
plt.xlabel("sugar level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.5,color='g')
```

```python
plt.xlabel("sugar level")
plt.ylabel("number of patients")
plt.title("Blood Sugar Chart")
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'])
plt.legend()
```
