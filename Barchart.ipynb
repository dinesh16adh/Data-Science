```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['google','amzn',' msft','fb']
revenue=[90,136,89,27]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue,label='Revenye',color='red')
plt.xticks(xpos,company)
plt.xlabel('software companies')
plt.ylabel('Revenye(Bin)')
plt.title('Us Technology Stocks')
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')
plt.xticks(xpos,company)
plt.ylabel('finance')
plt.title('Us Technology Stocks')
plt.legend()
```

```python
plt.bar(xpos,revenue,width=0.4,label="Revenue")
plt.bar(xpos,profit,width=0.4,label='Profit')
plt.xticks(xpos,company)
plt.ylabel('finance')
plt.title('Us Technology Stocks')
plt.legend()
```

```python
plt.barh(xpos,revenue,label='Revenue',color='red')
plt.yticks(xpos,company)
plt.xlabel('Revenye(Bin)')
plt.ylabel('Software Companies')
plt.title('Us Technology Stocks')
plt.legend()
```

```python
plt.barh(xpos-0.2,revenue,height=0.4,label="Revenue")
plt.barh(xpos+0.2,profit,height=0.4,label='Profit')
plt.yticks(xpos,company)
plt.xlabel('finance')
plt.title('Us Technology Stocks')
plt.legend()
```
