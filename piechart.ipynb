```python
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ['home rent','food','phone/internet bill',' car','other']
plt.pie(exp_vals,labels=exp_labels,autopct='%1.2f%%',rotatelabels=True,colors=['red','green','blue','orange','gray']);
plt.legend()
plt.title("hello")
```

```python
plt.pie(exp_vals,labels=exp_labels,autopct='%1.2f%%',rotatelabels=True,colors=['red','green','blue','orange'],explode=[0,0,0,0.1,0.2]);
```

```python
plt.pie(exp_vals,labels=exp_labels,autopct='%1.2f%%',rotatelabels=True,colors=['red','green','blue','orange','gray'],startangle=40);
```

```python
plt.pie(exp_vals,labels=exp_labels,autopct='%1.2f%%',rotatelabels=True,colors=['red','green','blue','orange','gray'],counterclock=True);
```
